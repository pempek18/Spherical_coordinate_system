```python
import numpy as np
import matplotlib.pyplot as plt 
import pylab as pl
# import sqrt from math module
from math import sqrt, pi
from Spherical import Spherical
```

```python
s = Spherical(False)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
arr = []
for i in np.arange(-pi, pi, 0.1):
    x = s.x(12, pi, i)
    y = s.y(12, pi, i)
    ax1.scatter(x=x, y=y)
    arr.append((x,y,i))
    # ax2.scatter(x=x, y = i)
    # ax2.scatter(x=y, y = i)

plt.title('Transfer spherical to xy', fontsize=11, pad=-20, loc='left')
plt.show()

data = np.array(arr)
# Extract x, y, and phi values for plotting
x_values = data[:, 0]   # x positions
y_values = data[:, 1]   # y positions
phi_values = data[:, 2] # phi values (angles)

# Plot y vs phi
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(phi_values, y_values, 'ro-', label='y vs φ')
plt.plot(phi_values, x_values, 'bo-', label='x vs φ')
plt.xlabel('Angle φ (degrees)')
plt.ylabel('xy position')
plt.title('xy vs φ')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Circle parameters
center_x, center_y, center_z = 0, 0, 0  # Center of the circle
radius = 5                               # Radius of the circle
num_points = 100                         # Number of points in the circle for smoothness
z_height = 2                             # Constant Z height for a flat circle in the XY plane

# Generate points for a circle in the XY plane with constant Z
phi = np.linspace(0, 2 * np.pi, num_points)  # Angle from 0 to 2π
x = center_x + radius * np.cos(phi)          # X coordinates
y = center_y + radius * np.sin(phi)          # Y coordinates
z = np.full(num_points, center_z + z_height)   # Z coordinates (constant height)

# Plotting the circular path
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='Circular path in XY plane', color='b')
ax.scatter(x[0], y[0], z[0], color='r', label='Start')
ax.scatter(x[-1], y[-1], z[-1], color='g', label='End')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title("XYZ Manipulator Running Circles in 3D Space")
ax.legend()

plt.show()
```
